# Problem Statement
Find a point in $\mathbb{R}^{2}$ with the least average Euclidean distance to a set of arbitrary points

```python
# First of all, we will import Random module from Python standard library
# To learn more about Random module:
# https://www.geeksforgeeks.org/python-random-module/
from random import Random

"""
We will define the seed and initialize the random generator
the seed is an initiale random value we define
to ensure the same random numbers will be generated whenever
we run the code at anytime
It is important to define a seed to ensure the model
works the same way everytime we run it
"""
SEED = 5
random_gen = Random(x = SEED)

"""
This function generates N random points in the unit square [0, 1] x [0, 1]
and returns their x-coordinates and y-coordinates as separate lists.

Parameters:
- N: int, optional (default=1000)
    The number of random points to generate.

Returns:
- list_x: list of floats
    A list of N random x-coordinates in the unit square.
- list_y: list of floats
    A list of N random y-coordinates in the unit square.

Algorithm:
- The function generates N random x-coordinates and y-coordinates independently,
using the uniform distribution over [0, 1] provided by the Python `random` module.
- The x-coordinates and y-coordinates are stored in separate lists, `list_x` and `list_y`.
- The lists are returned as a tuple (list_x, list_y).
"""

def generate_random_points(N = 1000):
  # random_gen = Random(x = SEED)
  
  """
  We start this function by initializing 2 empty lists, one for x-coordinates
  and another one for y-coordinates
  """
  list_x, list_y = [], []

  """
  First we will loop over range N (Default value is 1000)
  and fill the x-coordinates list with random numbers in range [0, 1]
  using the uniform method, to learn more about uniform methods:
  https://www.geeksforgeeks.org/python-number-uniform-method/?ref=lbp
  """
  for _ in range(N):
    list_x.append(random_gen.uniform(a = 0, b = 1))

  # Then we will loop over range N to fill the y-coordinates list
  for _ in range(N):
    list_y.append(random_gen.uniform(a = 0, b = 1))

  # We return x-coordinates and y-coordinates lists afterwards
  return list_x, list_y

data_x, data_y = generate_random_points()
print(data_x[:10])
print(data_y[:10])

# We will build the same function using list comprehension
# To learn more about list comprehension:
# https://www.geeksforgeeks.org/python-list-comprehension/

# We start by generating a new random function
random_gen = Random(x = SEED)

"""
This function generates N random points in the unit square [0, 1] x [0, 1]
and returns their x-coordinates and y-coordinates as separate lists,
using a list comprehension.

Parameters:
- N: int, optional (default=1000)
    The number of random points to generate.

Returns:
- list_x: list of floats
    A list of N random x-coordinates in the unit square.
- list_y: list of floats
    A list of N random y-coordinates in the unit square.

Algorithm:
- The function generates N random x-coordinates and y-coordinates independently,
using the uniform distribution over [0, 1] provided by the Python `random` module,
in two separate list comprehensions.
- The x-coordinates and y-coordinates are returned as a tuple (list_x, list_y).
"""

def generate_random_points_comp(N = 1000):
  # random_gen = Random(x = SEED)
  """
  We will fill the x-coordinates list by generating random numbers
  N times, then we will fill the y-coordinates list in the same way 
  """
  return (
      [random_gen.uniform(a = 0, b = 1) for _ in range(N)],
      [random_gen.uniform(a = 0, b = 1) for _ in range(N)],
  )

# A better way for the 2 methods above is to define the random generator inside
# each method of them, "Like in the commmented code lines inside each function"

data_x, data_y = generate_random_points_comp()
print(data_x[:10])
print(data_y[:10])
```

# The loss function
Let:
- The solution: ($x_{p}$, $y_{p}$)
- The dataset $\mathbb{D}$ is a {($x_{i}$, $y_{i}$)}
- $\mathbb{L}$ be the loss function and the solution be
$$(x_{p}, y_{p}) = argmin(\mathbb{L})$$
<br>
$$where\;\mathbb{L} = \frac{1}{N} \sum_{i=0}^{N-1}[(x_{i} - x_{p})^{2} + (y_{i} - y_{p})^{2}]^{\frac{1}{2}}$$

# Evaluate loss

```python
from math import sqrt

# Note that the doctor didn't make the loss function this way,
# he instead used the globally defined data_x, and data_y,
# but the best practice is to set them as parameters of the function

"""
Calculates the loss value between a point 
(x_p, y_p) and a set of data points (x_i, y_i) 
using the root mean squared distance between 
each data point and the point (x_p, y_p).

Parameters:
x_i (list of float): x-coordinates of the input data points
y_i (list of float): y-coordinates of the input data points
x_p (float): x-coordinate of the point to compare to the data points
y_p (float): y-coordinate of the point to compare to the data points

Returns:
loss (float): The root mean squared distance between
the point (x_p, y_p) and the data points
"""
def loss(x_i, y_i, x_p, y_p):
  # We first get the size of input list
  n = len(x_i)
  
  """
  We will use list comprehension to calculate the summation of the
  distance between the point (x_p, y_p) and all other points (x_i, y_i)
  then we will divide this summation by the size of (x_i, y_i) to return
  the loss as a float
  """
  return (1 / n) * sum(
      [sqrt((x_i - x_p)**2 + (y_i - y_p)**2)
        for x_i, y_i in zip(x_i, y_i)
      ]
  )

# This is an easier way to make the loss function
# without using list comprehension
def loss_easy(x_i, y_i, x_p, y_p):
  n = len(x_i)
  sum = 0
  for i in range(n):
    sum += sqrt((x_i[i] - x_p)**2 + (y_i[i] - y_p)**2)
  loss = sum / n
  return loss

print(loss(data_x, data_y, 5, 5))
print(loss_easy(data_x, data_y, 5, 5))
```

# Visualize loss

```python
# To visualize our loss function, we need to make a meshgrid
# we will set its length to 300, and each step in the grid has size of 0.01
STEP = 0.01
MESH_SIZE = 300

# Now, let's create our loss meshgrid, we will make it in a function
# This meshgrid is created by nested list comprehenshion
# To learn more about nested list comprehension:
# https://www.geeksforgeeks.org/nested-list-comprehensions-in-python/
def calculate_loss_mesh_comp(x_i, y_i, step = 0.01, mesh_size = 300):
  """
  First of all, we will create the x_mesh and y_mesh lists,
  they are 1D lists of x, y values in the meshgrid
  """
  x_mesh = [-1 + i * step for i in range(mesh_size)]
  y_mesh = [-1 + i * step for i in range(mesh_size)]
  
  """
  The loss mesh is a 2D list of loss values in the meshgrid,
  where the element at position (i, j) represents 
  the value of the loss function at the point (x_mesh[j], y_mesh[i])
  We will create an empty row for each x in the meshgrid
  and once the row is created, we will insert the loss value for each
  x and y in the x_mesh and y_mesh to this row, then we will go to
  create a new row
  """
  loss_mesh = [
    [loss(x_i, y_i, x_curr, y_curr) for x_curr in x_mesh] # row by row
      for y_curr in y_mesh # append list for each row
  ]
  return x_mesh, y_mesh, loss_mesh

# LET'S DO THE SAME WITHOUT LIST COMPREHENSION

"""
Creates a 2D meshgrid of the loss function over 
the square region [-1, 1] x [-1, 1].

Parameters:
x_i (list): A list of input x values for the loss function.
y_i (list): A list of corresponding y values for the loss function.
step (float): The spacing between adjacent points in the meshgrid, optional (Default: 0.01).
mesh_size (int): The number of points in each dimension of the meshgrid, optional (Default:300).

Returns:
x_mesh (list): A 1D list of x values in the meshgrid.
y_mesh (list): A 1D list of y values in the meshgrid.
loss_mesh (list): A 2D list of loss values in the meshgrid,
where the element at position (i, j) represents the value of the
loss function at the point (x_mesh[j], y_mesh[i]).
"""
def calculate_loss_mesh(x_i, y_i, step = 0.01, mesh_size = 300):
  # Create 1D lists of x and y values in the meshgrid
  x_mesh = []
  y_mesh = []
  for i in range(MESH_SIZE):
      x_mesh.append(-1 + i * STEP)
      y_mesh.append(-1 + i * STEP)

  # Create a 2D list of loss values in the meshgrid
  loss_mesh = []
  for i in range(MESH_SIZE):
      row = []
      for j in range(MESH_SIZE):
          x_curr = x_mesh[j]
          y_curr = y_mesh[i]
          loss_val = loss(x_i, y_i, x_curr, y_curr)
          row.append(loss_val)
      loss_mesh.append(row)
      
  # Return the resulting x, y, and loss lists as a tuple
  return x_mesh, y_mesh, loss_mesh

x_mesh, y_mesh, loss_mesh = calculate_loss_mesh(data_x, data_y)
```

# Numeric Conceptualization of Derivatives
It comes for the main definition of the derivative
<br><br>
$\frac{\partial \mathbb{L}}{\partial x} = \lim_{h \to 0} \frac{\mathbb{L}(x+h) - \mathbb{L}(x)}{h}$
<br><br>
By replacing the asymptotically infinitesimal by pragmatically small of values we can get a definition of deriviative that can be numerically computed 

```python
"""
Finds the minimum loss values for x and y parameters given a list of points (x_i, y_i)
and a starting point (x_p, y_p) using gradient descent optimization.

Parameters:
-----------
x_i: list
    A list of x values from the dataset to be used as reference points.
y_i: list
    The corresponding list of y values from the dataset to be used as reference points.
epochs: int
    The number of times to iterate through the optimization algorithm.
delta: float, default=0.01
    The learning rate for the gradient descent algorithm.
h: float, default=0.001
    The step size for approximating the gradient.
x_p: float, default=5
    The starting value for the x parameter in the optimization.
y_p: float, default=5
    The starting value for the y parameter in the optimization.
    
Returns:
--------
x_p: float
    The optimized value for the x parameter.
y_p: float
    The optimized value for the y parameter.
epoch_losses: list
    A list of the loss values for each epoch during optimization.
"""
def find_minimum(x_i, y_i, epochs, delta = 0.01, h = 0.001, x_p = 5, y_p = 5):
  epoch_losses = []
  for _ in range(epochs):
    epoch_losses.append(loss(x_i, y_i, x_p, y_p))

    """
    We will calculate the dloss/dx and dloss/dy
    using the mathmeatical formula to calculate the derivative
    """
    dloss_dx = (loss(x_i, y_i, x_p + H, y_p) - loss(x_i, y_i, x_p, y_p)) / h
    dloss_dy = (loss(x_i, y_i, x_p, y_p + H) - loss(x_i, y_i, x_p, y_p)) / h

    """
    We will then update both x_p and y_p values
    according to the new dloss/dx and dloss/dy
    """
    x_p -= delta * dloss_dx
    y_p -= delta * dloss_dy

  # Return x_, y_p and epoch_losses list after all epochs
  return x_p, y_p, epoch_losses

DELTA = 0.01
H = 0.001
x_p, y_p, epoch_losses = find_minimum(data_x, data_y, epochs = 3000, delta = DELTA, h = H)
x_p2, y_p2, epoch_losses2 = find_minimum(data_x, data_y, epochs = 1000, delta = DELTA, h = H)

# Let's visualize our losses curve for both 3000 epochs & 1000 epochs
fig_loss_curve = figure(figsize = (20, 10))
ax_loss_curve = fig_loss_curve.add_subplot(1, 2, 1)
ax_loss_curve2 = fig_loss_curve.add_subplot(1, 2, 2)
ax_loss_curve.plot(epoch_losses)
ax_loss_curve2.plot(epoch_losses2)
fig_loss_curve.show()

# Print the (x_p, y_p) for both 3000 epochs & 1000 epochs
print(x_p, y_p)
print(x_p2, y_p2)
```
